## Dataset #1: Traffic collisions - people and vehicles involved

```python
import pandas as pd
pd.options.mode.chained_assignment = None

original_df = pd.read_csv('pd_collisions_details_datasd.csv', low_memory=False)
beat = pd.read_csv('https://seshat.datasd.org/gis_police_beats/pd_beat_codes_list_datasd.csv').set_index('beat')
violation_section = pd.read_csv('violation_section.csv')
violation_extra = pd.read_csv('violations_extra.csv')
violation_section = pd.concat((violation_section, violation_extra), ignore_index = True)

beat_dict = beat.to_dict()['neighborhood']
collisions_sd = original_df.copy()
collisions_sd['date_time'] = pd.to_datetime(collisions_sd['date_time'])
collisions_sd['year'] = collisions_sd['date_time'].dt.year
collisions_sd['month'] = collisions_sd['date_time'].dt.month
collisions_sd['day'] = collisions_sd['date_time'].dt.day
collisions_sd['day_of_week'] = collisions_sd['date_time'].dt.day_of_week
collisions_sd['hour'] = collisions_sd['date_time'].dt.hour

collisions_sd['district'] = collisions_sd['police_beat'].apply(lambda x: beat_dict[x] if x in beat_dict else 'unknown')
collisions_sd['charge_desc'] = collisions_sd['charge_desc'].str.lower()

# Address
collisions_sd['street_address'] = (collisions_sd['address_no_primary'].astype(str)
                                   + collisions_sd['address_pd_primary']
                                   .apply(lambda x: x if x == ' ' else " " + x + " ")
                                     + collisions_sd['address_road_primary'] + " "
                                  + collisions_sd['address_sfx_primary'])
collisions_sd['street_address'] = collisions_sd['street_address'].str.strip().str.title().fillna("Unknown")
collisions_sd = collisions_sd.drop(['address_no_primary', 'address_pd_primary', 
                                    'address_road_primary', 'address_sfx_primary'], axis = 1)
collisions_sd['at_intersection'] = ((collisions_sd['address_pd_intersecting'] != ' ') 
                                    | (collisions_sd['address_name_intersecting'] != ' ') 
                                   | (collisions_sd['address_sfx_intersecting'] != ' '))
collisions_sd['street_address_intersection'] = (collisions_sd['address_pd_intersecting']
                                   + collisions_sd['address_name_intersecting'].astype(str)
                                   .apply(lambda x: x if x == ' ' else " " + x + " ")
                                     + collisions_sd['address_sfx_intersecting'])
collisions_sd['street_address_intersection'] = collisions_sd['street_address_intersection'].apply(
    lambda x: x if x != "   " else "None").fillna("Unknown")
collisions_sd['hit_run_lvl'] = collisions_sd['hit_run_lvl'].fillna("Not applicable").str.title()

collisions_sd = collisions_sd[
    (collisions_sd['person_role']=='DRIVER') &  # Filtering to only include DRIVER since we are interested in collision causes by driver
    (collisions_sd['year']>=2017)  # Filtering >= 2017 due to few observations in 2015 and 2016
]

collisions_sd = collisions_sd.merge(violation_section, how='left', left_on='violation_section', right_on='violation_section')

collisions_sd['person_injury_lvl'] = collisions_sd['person_injury_lvl'].fillna("UNCERTAIN")
collisions_sd['veh_type'] = collisions_sd['veh_type'].fillna("UNKNOWN")
collisions_sd['veh_make'] = collisions_sd['veh_make'].fillna("UNKNOWN")
collisions_sd['veh_model'] = collisions_sd['veh_model'].fillna("UNKNOWN")
collisions_sd['violation_section_description'] = collisions_sd['violation_section_description'].fillna('forbids drivers from stopping in a crosswalk and blocking it “unnecessarily.”')

collisions_sd.drop(['person_veh_type',
                    'person_role',
                    'address_pd_intersecting',
                    'address_name_intersecting',
                    'address_sfx_intersecting'], axis=1, inplace=True)
print('Shape of the df:', collisions_sd.shape)
collisions_sd.head(3)
```

The crux of the data here covers all vehicle accidents in San Diego from 2015 to 2024, April 19th. This data set is grouped to cover all types of brands and the collisions accidents they got into a singular dataframe. The most important sections will be brand and accident type (found under 'charge_desc') which will be the crux behind the true exploration of the data. Some other potentially useful columns we will use include the year and the location for exploration on aggregation on potential differences in the distribution based on the results we get from our main question. We also have an injured and killed count for further exploration into the severity of accidents.

A simple summary of the cleaning process is as follows; first all string formatting was lowered and title for better presentations. Some exceptions included model brand and type since we felt changin the string to lower wouldn't really do much in terms of presentations and the upper cases here helped highlight the specific type of brand, model, etc. of the vehicle. Police beats codes were replaced with the actual district number from the csv dictionary reference if it existed else unknown, and time reported was changed to datetime per pandas along with extracting the year in specific. Street addresses were concatenate to a single full name column and seperating columsn were dropped. The same was done on interesection street addresses if they existed along a column labeling whether the collision occured at any intersection or not. Finally the data was filtered to only driver roles due to interest in only collisions caused by drivers and reports from 2016 and 2015 were dropped due to extremly low amounts. Missing value columns were replaced appropiate will "missing" text or "not applicable" depending on the column context. Finally, merge was commited on violation_section with a seperate dataframe to include a description column on the specific violation.

### Subquestion 2: Are certain violation types more strongly associated with severe injuries or fatalities compared to others?

In our previous analysis of the datasets, we identified the types of violations that cause the most accidents. We also found that majority of accidents result in fewer than 3 injuries and have a very low lethality rate. We wanted to determine if there is a relationship between the type of violation and injuries/fatalities. For example, while red light violations are almost twice as common than drunk driving violations, is it possible that on average, drunk driving resulted in more injuries and fatalities? To answer this question, let's first do some descriptive analysis on average injuries and kill counts for different violations. We want to find out the mean aggregations for injuries on violations but we also don't want to have over-inflated values (for example violation '22650' has a mean injury of 4 but only one occurence). To do this we want to filter the data to only include violations that have occured at least 10 times within our dataset. Note it's good to remove these types of violations anyway since they would likely not be reoccuring enough to predict higher counts of injuries.

```python
hold = collisions_sd.groupby('violation_section').size()[collisions_sd.groupby('violation_section').size() > 10].index.tolist()
filted_viol = collisions_sd[collisions_sd['violation_section'].apply(lambda x: x in hold)]
viol_injuries = filted_viol.groupby('violation_section')['injured'].mean().to_frame().reset_index()
viol_injuries = viol_injuries.sort_values('injured', ascending=False)

plt.figure(figsize=(10, 6))
#For the sake of overplotting, we'll limit the means to those over 1.5
test = viol_injuries[viol_injuries['injured'] > 1.5]
mapping =  {
    '23153(G)': 'Driving under influence of alcohol and drugs',
    '23153(F)': 'Driving under influence of drugs',
    '21651(C)': 'Turn on divided highway or drive over divided section highway',
    '23153(A)': 'Driving under influence of alcohol',
    '14601(A)': 'Driving when under license suspension',
    '22405(A)': 'Driving under bridge or in tunnel over safe speed',
    '21460.5C': 'Not turning left on two left turn lane',
    '21453C':'Failure to yield to oncoming traffic during red light'
}
test['violation_section'] = test['violation_section'].apply(lambda x: mapping[x] if x in mapping else x)
sns.barplot(data=test, x='injured', y='violation_section', orient='h', color='cyan')
plt.ylabel('violation')
plt.xlabel('mean injured')
plt.yticks(fontsize=10)
plt.title('Mean injuries per violation section type')
plt.show()
```

Assuming the mean is not biased, driving under influence seems to be a cause more injuries on average in comparision to the other accidents as it has the largest mean under the grouping of "drugs and alcohol". Similarly the groupings of just alcohol and just drugs also have means that are similarly high

Let's repeat these steps now for death counts instead:

```python
filted_death = collisions_sd[collisions_sd['violation_section'].apply(lambda x: x in hold)]
viol_kills = filted_death.groupby('violation_section')['killed'].mean().to_frame().reset_index()
viol_kills = viol_kills.sort_values('killed', ascending=False)

plt.figure(figsize=(10, 6))
#For the sake of overplotting, we'll limit the means to those over 0.1,
#lower than injury count due to lower possible value range
test = viol_kills[viol_kills['killed'] > 0.1]
mapping =  {
    '23153(G)': 'Driving under influence of alcohol and drugs',
    '23153(F)': 'Driving under influence of drugs',
    '21456(C)': 'Driving despite a "walk" signal',
    '23153(A)': 'Driving under influence of alcohol',
    '21954A': 'Failure to yield vehicle for pedestrian',
    '21956': 'Pedestrian walking on road',
    '21209':'Driving vehicle on bike lane'
}
test['violation_section'] = test['violation_section'].apply(lambda x: mapping[x] if x in mapping else x)
sns.barplot(data=test, x='killed', y='violation_section', orient='h', color='cyan')
plt.yticks(fontsize=10)
plt.title('Mean deaths per violation section type')
plt.ylabel('violation')
plt.xlabel('mean killed')
plt.show()
```

Once again it seems driving under influence takes up a majority of these high means. Note that pedestrian related violations have appeared in the fatality plot but is absent from the injury plot.

So now that we have a better idea of how the distribution looks on violation type, let's do some testing. To start with we do can anova one-way testing just to confirm some differences in means exist, since we're doing multiple violation types on mean injury and kill counts. A significant score will tell us that there are statistically significant differences in mean injuries or mean kills between the difference violation types. We'll use the standard 0.05 threshold for our p-value. Remember since we're trying to avoid the inflated violation types on mean, we're choosing only violations with higher than 10 occurences!

```python
from scipy.stats import f_oneway

# Running anova one-way test between the different violation types

# Recall 'filted_viol' and 'filted_death' are the dataframes that contain violations with higher than 10 occurences. 
# (Note they're in reality the same dataframe, we simply store them in 2 different variables to avoid confusion)
injury_groups = ([filted_viol[filted_viol['violation_section'] == vt]['injured'] 
                  for vt in filted_viol['violation_section'].unique()])
death_groups = ([filted_death[filted_death['violation_section'] == vt]['killed'] 
                 for vt in filted_death['violation_section'].unique()])

f_stat_injury, p_value_injury = f_oneway(*injury_groups)
f_stat_death, p_value_death = f_oneway(*death_groups)
print("p_value for anova test of violation section types on injuries is " + str(p_value_injury))
print("p_value for anova test of violation section types on deaths is " + str(p_value_death))
```

It seems we got p-values that suggested there are significant mean differences of injuries and kill count between violation types but which specific groups do they differ for? We can use a posthoc test to find this out:

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# effectively what happens here is we're comparing mean for each violation against every other violation's mean 
# and seeing which results have significant differences! Then we store the post-hoc tests in a dataframe 
# for ease of pulling the data
tukey_injury = pairwise_tukeyhsd(endog=filted_viol['injured'], groups=filted_viol['violation_section'], alpha=0.05)
tukey_results = pd.DataFrame(data=tukey_injury.summary().data[1:], columns=tukey_injury.summary().data[0])

# Since we care about violations with high differences, we only look at places where we reject the null hypothesis.
significant_pairs = tukey_results[(tukey_results['reject'] == True)]

# We also care about violations that had the larger mean between the two comparisons,
# which is why we pull data from group1 or group2 depending on whether the meandiff was postive or negative.
sig1 = significant_pairs[significant_pairs['meandiff'] < 0].groupby('group1').size()
sig2 = significant_pairs[significant_pairs['meandiff'] > 0].groupby('group2').size()
number_of_sig_matches = sig1.add(sig2, fill_value=0)

# Once we've combined the two series, we have a list of violations that had rejected the null hypothesis 
# in at least one comparison AND were the larger mean in the comparison. If we're looking for violations that
# are associated with higher injury counts, we look for violations with higher means. Now we look at the violations
# that had the highest count of sig. larger mean differences against other violations. We had 116 unique violations 
# under the filtered data, so to find the ones with higher significant means, we'll look at the ones that were the 
# larger in comparisons on more than half of the total number of violations) Here's our results below:
number_of_sig_matches[number_of_sig_matches > 116/2]
```

These two violations look familar, let's try them again but match them to the mapping from our previous barplot:

```python
mapping =  {
    '23153(G)': 'Driving under influence of alcohol and drugs',
    '23153(F)': 'Driving under influence of drugs',
    '21456(C)': 'Driving despite a "walk" signal',
    '23153(A)': 'Driving under influence of alcohol',
    '21954A': 'Failure to yield vehicle for pedestrian',
    '21956': 'Pedestrian walking on road',
    '21209':'Driving vehicle on bike lane'
}
number_of_sig_matches[number_of_sig_matches > 116/2].set_axis([mapping['23153(A)'], mapping['23153(G)']])
```

So it seems the violations for driving under the influence of alcohol and 'alcohol and drugs' sees the highest mean in differences in the upper spectum, meaning there's statistical evidence that there's an association between these two violatons and higher injury counts.

Now let's do a similar process on the killed counts:

```python
# Same steps as we did before on injuries:

tukey_death = pairwise_tukeyhsd(endog=filted_death['killed'], groups=filted_death['violation_section'], alpha=0.05)
tukey_results = pd.DataFrame(data=tukey_death.summary().data[1:], columns=tukey_death.summary().data[0])

significant_pairs = tukey_results[(tukey_results['reject'] == True)]

sig1 = significant_pairs[significant_pairs['meandiff'] < 0].groupby('group1').size()
sig2 = significant_pairs[significant_pairs['meandiff'] > 0].groupby('group2').size()
number_of_sig_matches = sig1.add(sig2, fill_value=0)

# The one main difference is we should change the amount of matches we want to find for larger significance since
# the data range on killed count is much smaller. Let's do an 80% or higher find instead of the 50% on 
# injuries previously
number_of_sig_matches[number_of_sig_matches > (116 * 0.8)]
```

```python
# Using the mapping to confirm what these violations are:
mapping =  {
    '23153(G)': 'Driving under influence of alcohol and drugs',
    '23153(F)': 'Driving under influence of drugs',
    '21456(C)': 'Driving despite a "walk" signal',
    '23153(A)': 'Driving under influence of alcohol',
    '21954A': 'Failure to yield vehicle for pedestrian',
    '21956': 'Pedestrian walking on road',
    '21209':'Driving vehicle on bike lane'
}
number_of_sig_matches[number_of_sig_matches > (116 * 0.8)].set_axis(
    [mapping['21456(C)'], mapping['21954A'], mapping['23153(F)']])
```

Thus we can say that there's evidence for association between higher death counts and the above violations. Now we can answer our subquestion.

- The data could negatively impact the car sales of brands that we find to get into more car accidents.
- Bad publicity from the study could cause affected car brands to make unnecessary changes to their products to include extra safety features which would add to the price of the car.
- It could instill fear among drivers in San Diego, impacting their perception of certain brands.
- Mainly the idea of brand labeling as negative unintentionally when we're just trying to explore this potential relationship to help clarify whether it's relevant or even exists.
- Whatever conclusion we reach that is supported by the data needs to be as unbiased as possible, else we risk negative stereotyping as mentioned in our potential ethical concerns i.e. no negative connotations to a specific group of drivers like Tesla drivers etc.
- Young drivers (aka worse drivers) causing accidents with safe car brands could cause cars with better safety features to be misconstrued as unsafe.
- Insurance companies could use the study as justification to raise insurance premiums on those car brands.
- Bias issues will largely be labeled by disclaimers. If we can avoid it, we will but if it's too integrated with the data, the next step is pinpoint and point out this fact to an audience and clarify it to avoid potential bias forming because of it
